```python
class TiendaOnline: #creamos la clase, en mayúscula la 1 letra de cada palabra.
    def __init__(self): # creamos el método constructor, donde definimos los atributos. self siempre debe estar en lso atributos,ya que lo vamos ausar para llamar a nuestro objeto.
        self.inventario = []
        self.ventas_totales = 0.0
        self.clientes = {}
    def agregar_producto(self, nombre, precio,cantidad):#creamos los métodos(propiedades que le vamos a dar a nuestro objeto), siempre nombramos los atributos.
        for i in self.inventario: # accedemos a cada elemento de la lista del inventario.
            if i["nombre"] == nombre: #miramos a ver que exista el producto.
                i["cantidad"] = i["cantidad"]+ cantidad  #añadimos una cantidad nueva en caso de que que el producto exista, se suma a la existente.
                print(f"El producto {nombre} ha sido actualizado con una cantidad de {cantidad} unidades")
                return
        #si no existe añado el producto
        nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad} #creo la variable nuevo producto.
        self.inventario.append(nuevo_producto) #añado a la lista de inventario el producto.
        print(f"El producto {nuevo_producto["nombre"]} ha sido añadido correctamente. Con una cantidad de {nuevo_producto["cantidad"]} unidades y un precio de {nuevo_producto["precio"]} euros")
        
    def ver_inventario(self): #creamos método con los parámetros.
        for i in self.inventario: #recorremos con for cada elemento de la lista
            print(f" El inventario disponible es Nombre: {i["nombre"]}, Precio: {i["precio"]} euros, Cantidad: {i["cantidad"]}") #hacemos un print que nos devuelva cada elemento.
        return
        
    def buscar_producto(self, nombre): #creamos método con los parámetros.
        for i in self.inventario: #recorremos con for cada elemento de la lista.
            if i["nombre"] == nombre: #si el nombre del elemento dentro de nuestra lista coincide con el nombre proporcionado.
                print(f" El producto que buscas se encuentra en nuestro catálogo, aquí tienes los detalles: Nombre: {i["nombre"]}, Precio: {i["precio"]} euros, Cantidad: {i["cantidad"]}") #le decimos que nos devuelva el producto.
                return
     
        print("Lo sentimos, el producto que buscas no se encuentra en nuestro catálogo.") #y si no que nos diga que no está.
        
    def actualizar_stock(self, nombre, cantidad):  #creamos método con los parámetros.
        for i in self.inventario: #recorremos con for cada elemento de la lista.
            if i["nombre"] == nombre: #hacemos un condicional para saber si el nombre del producto se encuentra en la lista, según el parámetro que nos pasen (si son iguales).
                i["cantidad"] = i["cantidad"] + cantidad #añadimos la cantidad que nos da el usuario y sumamos o restamos a la existente.
                print(f"La cantidad del producto ha sido actualizada, ahora es de: {i["cantidad"]} unidades") #devolvemos mensaje satisfactorio.
                return
        print("Lo sentimos, no tenemos dispoblible ese producto") #devolvemos mensaje en caso de que el producto no este. 
    
    def eliminar_producto(self, nombre):#creamos método con los parámetros.
        for i in self.inventario: #recorremos con for cada elemento de la lista.
            if i["nombre"].lower() == nombre.lower(): #si el nombre del elemento dentro de nuestra lista coincide con el nombre proporcionado. le pasamos la funcion a miniscula, para que al llamar a la instancia me coja todo.
                self.inventario.remove(i) #le indicamos que elimine el elemento de la lista.
                print(f"El producto {i["nombre"]}, ha sido eliminado correctamente") #nos diga que ha sido eliminado con exito e imprimos nombre.
                return
        print("¡Vaya! parece ser que el producto no se encuentra en el inventario") #si no se encuentra, que nos diga que no está.
        
    def calcular_valor_inventario(self): #creamos método con los parámetros.
        valor_total_inventario = 0.0 #creamos un contador para guardar la suma de los productos, empieza en 0.
        for i in self.inventario:#recorremos con for cada elemento de la lista.
            precio = i["precio"] #accedo al precio y creo variable.
            cantidad = i["cantidad"]#accedo a la cantidad y creo variable.
            valor_total_inventario = valor_total_inventario + (precio*cantidad) #Sumamos el valor del producto actual al total acumulado
        print(f"El valor total del inventario es de: {valor_total_inventario} euros")
        return

    def agregar_clientes(self, nombre, email): #creamos método con los parámetros.
        if nombre in self.clientes: ##verificamos si el cliente existe. Utilizando in
            print(f"El cliente con nombre: {nombre} ya existe")
            return
        self.clientes[nombre] = {'email': email,'compras': []} # si no existe, accedemos al diccionario(indexamos por nombre "clave") y sus respectivos valores "email y "compras".
        print(f" El cliente con nombre: {nombre} e email: {email} ha sido añadido correctamente") #damos un mensaje de que el cliente ha sido añadido con éxito.
       
    def ver_cliente(self): #creamos método con los parámetros.
        for nombre, datos in self.clientes.items(): #recorremos el diccionario con un for para ver los clientes. al ser un diccinario tenemos que ver clave(nombre)-valor(dato). usamos. items para que nos lo empaquete (clave-valor) y nos devuleva el pack ya que si iterasemos sin ello son nos daria los valores. 
            print(f"la lista de clientes registrados es: Nombre: {nombre}, Email: {datos["email"]}, Compras: {datos["compras"]}") #le decimos que nos devuelva el diccionario de clientes.
        return
    def realizar_compra(self): #creamos método con los parámetros.
        carrito = [] #creamos una variable para almacenar los productos.
        coste_total_compra = 0.0 #otra para las compras totales.
        
    #pedimos al usuario que se registre.
        nombre_usuario = input("por favor escribe tu nombre, para realizar la compra")
        email_usuario = input("Por favor escribe tu email")
        if nombre_usuario in self.clientes:
            print("Ya estás registrado, puedes proceder a tu compra") #si esta registrado mensaje de ok.
        else: 
            self.agregar_clientes(nombre_usuario, email_usuario) #agregamos el usario a nuestro diccionario en caso de que este.
            print("¡Es hora de comprar!")
               
        while True: # Hacemos un bucle while para que se puedan realizar la compra varias veces y dentro operamos.

            print("\n--- INVENTARIO DISPONIBLE ---")
            self.ver_inventario() #sacas neustro inventario disponible llamando a la función.
            
            solicitud_producto = input(f"Por favor, seleccione el producto que desea comprar") #input solicitando producto.
        
            cantidad_deseada = int(input("Ahora que has seleccionado el producto danos la cantidad  que deseas en número"))    
            
            for i in self.inventario: #iteramos por cada elemento del inventario para ver si el producto que el cliente ha elegido se encuentra en stock.
               
                if i["nombre"].lower() == solicitud_producto.lower(): #verificamos que este el producto.
                    
                    if cantidad_deseada <= i["cantidad"]: # ahora la cantidad.
                        nuevo_producto = {"nombre": solicitud_producto, "precio": i['precio'], "cantidad": cantidad_deseada} #creamos variale para guardar los productos y añadir al carrito
                        carrito.append(nuevo_producto) #añadimos al carrito.
                        self.actualizar_stock(solicitud_producto,-cantidad_deseada) #actualizamos stock, llamando a la función.
                        coste_total_compra = coste_total_compra + i["precio"] * cantidad_deseada #actualziamos el coste de la compra.
                        print(f"Tenemos el producto: {solicitud_producto}. ¡ Compra realizada!") 
                        print(f"En tu carrito tienes los siguientes productos: {carrito}")
                        print(f"llevas un coste total de {coste_total_compra} euros")
                        break #se para y rompe el bucle
                    else: 
                        print(f"Lo sentimos para el producto {solicitud_producto}, que intentas comprar no hay suficiente stock, vuleva a intentarlo")
                        break #salimos del for,vuelve al inicio del while.
            else:
                print(f"El producto que has seleccionado: {solicitud_producto} no fue encontrado en el catálogo, por favor intentalo de nuevo.") #si no está el producto salta diciendo que no existe y vuelve al while.
            
            continuar_compra = input("¿Desea añadir otro producto a su carrito? (escriba 's' para sí o 'n' para no): ").strip().lower()
            
            if continuar_compra == 'n':
                break  #así rompo el bucle while si no querio seguir comprando.
                  
        print(f"¡Compra finalizada! El coste total de tu compra es de:{coste_total_compra} euros y llevas estos elementos en el carrito: {carrito}.")
              
    
```
